# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
district_df=school_data_complete.groupby(["school_name"])

total_schools_list = district_df["school_name"].count()
total_schools=total_schools_list.count()

total_budget_schools = district_df.max()
total_budget=total_budget_schools["budget"].sum()

total_students = school_data_complete["student_name"].count()
average_math_score = school_data_complete["math_score"].mean()
average_reading_score = school_data_complete["reading_score"].mean()
num_passing_math = school_data_complete.loc[school_data_complete["math_score"] > 69].count()["student_name"]
percent_passing_math = (num_passing_math / total_students)
num_passing_reading = school_data_complete.loc[school_data_complete["reading_score"] > 69].count()["student_name"]
percent_passing_reading = (num_passing_reading / total_students)
percent_overall_passing = (average_math_score + average_reading_score) / 2/100

district_summary = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students], "Total Budget": [total_budget], "Average Math Score": [average_math_score], "Average Reading Score": [average_reading_score], "% Passing Math": [percent_passing_math], "% Passing Reading": [percent_passing_reading], "% Overall Passing Rate": [percent_overall_passing]})

district_summary["Total Students"]=district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"]=district_summary["Total Budget"].map("${:,}".format)
district_summary["Average Math Score"]=district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"]=district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["% Passing Math"]=district_summary["% Passing Math"].map("{:.2%}".format)
district_summary["% Passing Reading"]=district_summary["% Passing Reading"].map("{:.2%}".format)
district_summary["% Overall Passing Rate"]=district_summary["% Overall Passing Rate"].map("{:.2%}".format)

district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name

```python
# School Size
school_size = pd.DataFrame(district_df.size())
school_size = school_size.rename(columns={0:"Size"})
```

```python
# School Budget
school_budget=pd.DataFrame(total_budget_schools["budget"])
```

```python
# Math and Reading Avg
school_avg_scoresall=district_df.mean()
school_avg_scores=school_avg_scoresall[["reading_score","math_score"]]
```

```python
#School Math Passing Rate
num_passing_math_total = school_data_complete[school_data_complete["math_score"] > 69]
num_passing_math_school=num_passing_math_total.groupby("school_name")
number_pass_math = num_passing_math_school.count()
number_pass_math = pd.DataFrame(number_pass_math["math_score"])
number_pass_math = number_pass_math.rename(columns={"math_score":"number students pass math"})
per_pass_math_num=pd.merge(number_pass_math,school_size,how='left',on="school_name")
per_pass_math_num
per_pass_math=pd.DataFrame(per_pass_math_num["number students pass math"]/per_pass_math_num["Size"])
per_pass_math = per_pass_math.rename(columns={0:"% Passing Math"})
```

```python
#School Reading Passing Rate
num_passing_reading_total = school_data_complete[school_data_complete["reading_score"] > 69]
num_passing_reading_school=num_passing_reading_total.groupby("school_name")
number_pass_reading = num_passing_reading_school.count()
number_pass_reading = pd.DataFrame(number_pass_reading["reading_score"])
number_pass_reading = number_pass_reading.rename(columns={"reading_score":"number students pass reading"})
per_pass_reading_num=pd.merge(number_pass_reading,school_size,how='left',on="school_name")
per_pass_reading_num
per_pass_reading=pd.DataFrame(per_pass_reading_num["number students pass reading"]/per_pass_reading_num["Size"])
per_pass_reading = per_pass_reading.rename(columns={0:"% Passing Reading"})
```

```python
#School Type
school_type=school_data[["school_name","type"]]
school_type=school_type.set_index("school_name")
```

```python

school_sum=pd.merge(school_type,school_size,on="school_name")
school_sum=pd.merge(school_sum,school_budget,on="school_name")
school_sum["Per Student Budget"]=school_sum["budget"]/school_sum["Size"]
school_sum=pd.merge(school_sum,school_avg_scores,on="school_name")
school_sum=pd.merge(school_sum,per_pass_reading,on="school_name")
school_sum=pd.merge(school_sum,per_pass_math,on="school_name")
school_sum["% Overall Passing Rate"]=(school_sum["% Passing Reading"]+school_sum["% Passing Math"])/2

school_sum2=pd.merge(school_type,school_size,on="school_name")
school_sum2=pd.merge(school_sum2,school_budget,on="school_name")
school_sum2["Per Student Budget"]=school_sum2["budget"]/school_sum2["Size"]
school_sum2=pd.merge(school_sum2,school_avg_scores,on="school_name")
school_sum2=pd.merge(school_sum2,per_pass_reading,on="school_name")
school_sum2=pd.merge(school_sum2,per_pass_math,on="school_name")
school_sum2["% Overall Passing Rate"]=(school_sum2["% Passing Reading"]+school_sum2["% Passing Math"])/2

school_sum2 = school_sum2.rename(columns={"type":"School Type","Size":"Total Students","budget":"Total School Budget","reading_score":"Average Reading Score","math_score":"Average Math Score"})
school_sum2["Total Students"]=school_sum2["Total Students"].map("{:,}".format)
school_sum2["Total School Budget"]=school_sum2["Total School Budget"].map("${:,}".format)
school_sum2["Per Student Budget"]=school_sum2["Per Student Budget"].map("${:,.2f}".format)
school_sum2["Average Reading Score"]=school_sum2["Average Reading Score"].map("{:.2f}".format)
school_sum2["Average Math Score"]=school_sum2["Average Math Score"].map("{:.2f}".format)
school_sum2["% Passing Reading"]=school_sum2["% Passing Reading"].map("{:.2%}".format)
school_sum2["% Passing Math"]=school_sum2["% Passing Math"].map("{:.2%}".format)
school_sum2["% Overall Passing Rate"]=school_sum2["% Overall Passing Rate"].map("{:.2%}".format)

school_sum2
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Filter by grade
ninth_grade_all = school_data_complete.loc[school_data_complete["grade"] == "9th"]
tenth_grade_all = school_data_complete.loc[school_data_complete["grade"] == "10th"]
eleventh_grade_all = school_data_complete.loc[school_data_complete["grade"] == "11th"]
twelth_grade_all = school_data_complete.loc[school_data_complete["grade"] == "12th"]

#Create series by grade of average score
ninth_math_sum = ninth_grade_all.groupby("school_name").mean()
ninth_math_sum = pd.DataFrame(ninth_math_sum["math_score"])
ninth_math_sum = ninth_math_sum.rename(columns={"math_score":"9th"})

tenth_math_sum = tenth_grade_all.groupby("school_name").mean()
tenth_math_sum = pd.DataFrame(tenth_math_sum["math_score"])
tenth_math_sum = tenth_math_sum.rename(columns={"math_score":"10th"})

eleventh_math_sum = eleventh_grade_all.groupby("school_name").mean()
eleventh_math_sum = pd.DataFrame(eleventh_math_sum["math_score"])
eleventh_math_sum = eleventh_math_sum.rename(columns={"math_score":"11th"})

twelth_math_sum = twelth_grade_all.groupby("school_name").mean()
twelth_math_sum = pd.DataFrame(twelth_math_sum["math_score"])
twelth_math_sum = twelth_math_sum.rename(columns={"math_score":"12th"})

math_sum=pd.merge(ninth_math_sum,tenth_math_sum,on="school_name")
math_sum=pd.merge(math_sum,eleventh_math_sum,on="school_name")
math_sum=pd.merge(math_sum,twelth_math_sum,on="school_name")

math_sum
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#Create series by grade of average score
ninth_reading_sum = ninth_grade_all.groupby("school_name").mean()
ninth_reading_sum = pd.DataFrame(ninth_reading_sum["reading_score"])
ninth_reading_sum = ninth_reading_sum.rename(columns={"reading_score":"9th"})

tenth_reading_sum = tenth_grade_all.groupby("school_name").mean()
tenth_reading_sum = pd.DataFrame(tenth_reading_sum["reading_score"])
tenth_reading_sum = tenth_reading_sum.rename(columns={"reading_score":"10th"})

eleventh_reading_sum = eleventh_grade_all.groupby("school_name").mean()
eleventh_reading_sum = pd.DataFrame(eleventh_reading_sum["reading_score"])
eleventh_reading_sum = eleventh_reading_sum.rename(columns={"reading_score":"11th"})

twelth_reading_sum = twelth_grade_all.groupby("school_name").mean()
twelth_reading_sum = pd.DataFrame(twelth_reading_sum["reading_score"])
twelth_reading_sum = twelth_reading_sum.rename(columns={"reading_score":"12th"})

reading_sum=pd.merge(ninth_reading_sum,tenth_reading_sum,on="school_name")
reading_sum=pd.merge(reading_sum,eleventh_reading_sum,on="school_name")
reading_sum=pd.merge(reading_sum,twelth_reading_sum,on="school_name")

reading_sum
```
